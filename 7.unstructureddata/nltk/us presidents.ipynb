Every 4 years a new US president is sworn in and as a consequence has the right to hold an inaugural speech. These speeches have been made available in the nltk toolkit.

```python
import nltk
from nltk.corpus import inaugural
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

The first five speeches were

```python
fileids = inaugural.fileids()
print(fileids[0:5])
```

The year of the speech is always in the beginning of the text. The first 4 digits can be retrieved using  `list comprehension`.

```python
print([fileid[:4] for fileid in inaugural.fileids()][-5:])
```

How many times is the word "war"  and "america" used ?
Group the outcome into a list "collection"

```python
collection = [(target,fileid[:4]) for fileid in inaugural.fileids() 
 for w in inaugural.words(fileid) 
 for target in ['war', 'america'] if w.lower().startswith(target)]
collection[0:5]
```

Evolution of the words war and america

```python
cfd = nltk.ConditionalFreqDist(collection)
fig  = plt.figure(figsize=(15,8))
cfd.plot();
```
